```python
# Package imports
import seaborn as sns
import pandas as pd
import missingno as msno
```

### Ejemplo de uso

Vamos a cargar los datos de titanic donde ya sabemos que existe algunas columnas con datos nulos

Los graficos que utilizaremos son 3:

Bar plot
Matrix Plot
HeatMaps

```python
#load titanic from seaborn datasets pre load
df_titanic = sns.load_dataset("titanic")
```

### Bar Plot

Verifica en todas las columnas la cantidad de valores no nulos , y alli podemos identificar cuales son las barras que no llegan al tope , indicando cuales son las columnas que tiene datos nulos y su cantidad

```python
#Create bar plot
msno.bar(df_titanic);

# me muestra la cantidad de nulos por columna
#por ejemplo en age hay que rellenar el 20% de los datos
```

### Matrix Plot

En la Matrix , lo que hace es poder no solo indentificar que columnas tiene datos nulos , sino aproximadamente su ubicacion dado que se imagina que son lineas donde cada linea tiene los datos

Los espacios en blanco son los lugares de la matrix de datos , donde no hay datos nulos

Facil para poder identificar si hay observaciones donde hay una relacion de m,uchos datos nulos

```python
#Crate matrix plot
msno.matrix(df_titanic)

# podemos ver una correlacion inversa entre deck y age (cuando hay una, la otra no esta)
# tambien vemos cierta correlacion entre embarked y embark_town


#ver la relacion entre los nulos
```

### Heat Map

Aqui solo toma las columnas que tiene datos nulos y hace una correlacion entre ellas , para verificar si las columnas que existen son las mismas filas donde los valores son nulos

Ejemplo aqui de correlacion positiva es la variable Embarked y embarked town

```python
msno.heatmap(df_titanic)

# me muestra la correlacion entre las columnas con los datos nulos
#podemos ver que columnas son las que hay que borrar
#superior a 0.6 es correlacion entonces vemos que tengo que borrar la columna embarkes o la embark_town porque salr 1 de corralcion
#muestra solo la interaccion de las que presentaron algun nulo
```
