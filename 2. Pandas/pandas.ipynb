### Nan --> NOT A NUMBER
Si los indices no coinciden me llena las columnas con numeros vacios: Nan
- Los algoritmos de machine learning no pueden tener numeros Nan (ceros)
- Los Nan distorcionan los datos

FORMAS DE MANIPULARLOS
- isNull (si es nulo marca true)
- sum (cuenta la cantidad de nulos)
- dropna (elimina las filas que contienen nulos)

```python
tabla.isnull() #devuelve true el dato que en null, tambien puede ser isna (que es de numoy)
```

```python
tabla.isnull().sum() #cuenta la cantidad de nulos
```

```python
tabla.fillna(0) #rellena con cero
```

```python
tabla.dropna()  #borra la fila que tiene nulos
```
