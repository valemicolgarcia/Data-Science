```python
fig, ax = plt.subplots()
ax.plot([1, 3], [2, 4])
ax.set_ylim([2,4]) #limite y
ax.set_xlim([1,3]) #limite x
ax.set_xlabel('x') #titulo de la x
ax.set_ylabel('y') #titulo de la y
ax.set_title('Grafico x vs y') #Titulo general del grafico
ax.axhline(y=3, xmin=0, xmax=4, color='y') #linea horizontal  en yellow
ax.axvline(x=2.0, ymin=0, ymax=10, color='r') #linea vertical  en red
```

```python
df_lluvias = pd.read_csv('C:\\Users\\VICTUS\\Documents\\2024\\DATA SCIENCE\\7. MatPlotLib\\pune_1965_to_2002.csv')
df_lluvias.head()
```

```python
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year', axis='columns')  #borro la columna year
df_lluvias
```

```python
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
y
```

```python
fig, ax = plt.subplots(figsize=(12, 4)) #seteo el lienzo
ax.plot(x, y, label='Precipitaciones acumuladas') #le paso dos variables, label es lo que represdenta la linea
ax.set_xlabel('Año') #titulo del eje x
ax.set_ylabel('Precipitación acumulada (mm.)') #titulo del eje y
ax.set_title('Precipitaciones acumuladas según el año') #titulo del grafico
ax.legend() #haces que aparezca el label precipitaciones acumuladas, ademas podes agregarle especificaciones en algun recuadro
```

```python
fig, ax = plt.subplots() #seteo el lienzo
mapeo_colores = ax.scatter(df_lluvias['Aug'], df_lluvias['Sep'], c=df_lluvias.index) #c indica color, pide que los saque segun el indice (le da una escala de color arbitraria de la libreria)
fig.colorbar(mapeo_colores) #mapea los colores sobre esta figura y hace la barra al costado que indica el degrade junto con el nro de indice
ax.set_title('Precipitaciones Agosto-Septiembre')
ax.set_xlabel('Precipitaciones en Agosto (mm.)')
ax.set_ylabel('Precipitaciones en Septiembre (mm.)')
```

GRAFICOS DE BARRAS

```python
fig, ax = plt.subplots(figsize=(8,4)) #seteo el lienzo
precipitaciones_acumuladas = df_lluvias.sum() 
ax.bar(df_lluvias.columns, precipitaciones_acumuladas) #bar --> grafico de barras
ax.set_title('Precipitaciones acumuladas desde 1965 a 2002, según el mes')
ax.set_ylabel('Precipitación total (mm.)')
ax.set_xlabel('Mes')
```

HISTOGRAMA --> sirve para saber la distribucion del dato
- para saber el comportamiento de la variable

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df_lluvias.values.flatten(), bins=10) #histograma, columna x columna o variable x variable
ax.set_title('Histograma de precipitaciones')
ax.set_xlabel('Intervalos de precipitaciones (mm.)')
ax.set_ylabel('Frecuencia absoluta')
```

DIAGRAMA DE CAJAS

```python
fig, ax = plt.subplots(figsize=(8, 4))
#categorica vs continua
ax.boxplot(x=df_lluvias.T,)
ax.set_title('Boxplot de precipitaciones')
ax.set_xlabel('Meses')
ax.set_ylabel('Precipitacion (mm)')
#valores atipicos --> outliers
```

```python
fig, ax = plt.subplots(figsize=(12, 3)) #seteo el lienzo
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero') 
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1') #color c1 lo pone en rojo para distinguir
#los dos graficos los hace en el mismo lienzo porque lo indique solo una vez
maximo_enero = df_lluvias['Jan'].max() #traigo el maximo del dataframe
maximo_febrero =  df_lluvias['Feb'].max()
ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5,    linewidth=3, label='Máxima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='Máxima de febrero')
ax.set_xlabel('Año')
ax.set_ylabel('Precipitación (mm.)')
ax.set_title('Precipitaciones de enero y febrero')
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1]) #hasta donde quiero que sea la x, -1 es el final
ax.grid = True
ax.titleweight = 'normal'
ax.legend()
```
