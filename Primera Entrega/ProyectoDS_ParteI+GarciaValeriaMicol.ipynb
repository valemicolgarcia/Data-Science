VARIABLES CATEGÓRICAS --> object

VARIABLES NUMÉRICAS --> float64 --> considero que estas son las variables de mayor relevancia

- Key: La clave en la que se encuentra la pista
- Loudness: El volumen general de la pista en decibeles (dB)
- Speechiness: Detecta la presencia de palabras habladas en la pista
- Acousticness: Una medida de confianza para saber si la pista es acustica
- Instrumentalness: Indica si una pista no contiene voces
- Liveness: Detecta la presencia de una audiencia en la grabación
- Valence: Describe la positividad musical que transmite una pista
- Tempo: El tempo general estimado en pulsaciones por minuto (BPM)
- Duration_ms: Duración de la canción en milisegundos
- Views: Visualizaciones en Youtube
- Likes: Likes en Youtube
- Comments: Comentarios en Youtube
- Stream: Número de reproducciones de la canción en Spotify

# Visualizaciones 

### 1. Hay relación entre alguna característica musical y la popularidad?

RELACIÓN ENTRE LA ACÚSTICA DE LA CANCION Y LA CANTIDAD DE VISUALIZACIONES

- Grafico de Dispersión con Bokeh

```python
from bokeh.palettes import Spectral6
from bokeh.models import HoverTool
from bokeh.io import show, output_notebook
output_notebook()
from bokeh.embed import file_html
from bokeh.models import CategoricalColorMapper, ColumnDataSource, FactorRange
from bokeh.plotting import figure, output_file
```

```python
# Seleccion de datos
from bokeh.models import Range1d
x_scatter = df['Acousticness']# Data en x
y_scatter = df['Views'] # data en y

scatter_plot = figure(width=700, height=300, x_axis_label='Acousticness', y_axis_label='Views', title='Acousticness vs Views') # Grafico (Fondo)
scatter_plot.circle(x_scatter, y_scatter, size=5, line_color='navy', fill_color='orange', fill_alpha=0.5) # Scatter plot
scatter_plot.add_tools(HoverTool()) # Agrego opcion interactiva
scatter_plot.y_range= Range1d(7400, 8800) # Definir limites
scatter_plot.grid.visible= False


show(scatter_plot)
```

Podemos observar que los datos no tienen una relación entre si, es decir que la acústica no esta linealmente relacionada con la cantidad de visualizaciones

RELACIÓN ENTRE LA DURACIÓN DE LA CANCIÓN Y LA CANTIDAD DE VISUALIZACIONES

- Gráfico de puntos o de dispersión con MatPlotLib

```python
fig, ax = plt.subplots()
mapeo_colores = ax.scatter(df['Duration_ms'], df['Views'], c=df.index) 

fig.colorbar(mapeo_colores)
ax.set_title('Duration vs Views')
ax.set_xlabel('Duration (ms)')
ax.set_ylabel('Views')

ax.grid(False)
```

Podemos observar que la duración y la cantidad de visualizaciones no tienen una relacion lineal positiva, sino que identificamos puntos que se distribuyen al azar sin un patrón obvio (falta de dependencia entre los valores)

### 2. Hay relación entre likes, comentarios y visualizaciones?

- Gráfico de puntos o de dispersión con SeaBorn

```python
f, axs = plt.subplots(1, 3, figsize=(15, 6), gridspec_kw=dict(width_ratios=[3, 3, 3])) # seteo el lienzo

sns.scatterplot(data=df, x='Views', y='Comments', hue='Album_type', ax=axs[0])
axs[0].set_title('Views vs Comments')
sns.scatterplot(data=df, x='Views', y='Likes', hue='Album_type', ax=axs[1])
axs[1].set_title('Views vs Likes')
sns.scatterplot(data=df, x='Comments', y='Likes', hue='Album_type', ax=axs[2])
axs[2].set_title('Comments vs Likes')
f.tight_layout()
```

Podemos observar que estas variables estan intimamente relacionadas, al haber mas visualizaciones hay mas likes y comentarios, el segundo gráfico es el que se observa con mayor claridad

Grafico que incluye las tres variables a relacionar: LIKES, COMENTARIOS, VISUALIZACIONES

Estan super relacionadas estas tres variables, al aumentar las visualizaciones, aumentan los likes y los comentarios

VIEWS, LIKES Y COMENTARIOS

```python
ax = sns.scatterplot(data=df, x='Views', y='Comments', hue='Likes')
ax.grid(False)
ax.set_title('Views, Comments, Likes')
```

### 6. Cómo es la distribución de mis datos?

- DISTRIBUCIÓN DE LA DURACIÓN, HISTOGRAMA CON BOKEH

```python

data = df['Duration_ms']
data=data[~np.isnan(data)]
hist, edges = np.histogram(data, density=True, bins=40)
p = figure(height=400,width=400, title='Distribution of Duration')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="white", fill_color='orange')
p.xaxis.axis_label = "Duration"
p.yaxis.axis_label = "Frecuency"
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.add_tools(HoverTool())
show(p)
```

La distribución esta sesgada hacia la derecha, con un pico en torno a las duraciones mas cortas

- DISTRIBUCIÓN DE LA ACÚSTICA, HISTOGRAMA CON SEABORN

```python
ax = sns.histplot(data=df, x = 'Acousticness', hue='Album_type')
plt.title('Distribution of Acousticness')
ax.grid(False)
```

La distribución parece bastante uniforme, con pico en los valores mas bajos. Hay una proporción mas alta de canciones con mayor acústica

- DISTRIBUCIÓN DE LA VALENCIA EN LAS CANCIONES, HISTOGRAMA CON SEABORN

- DISTRIBUCIÓN DEL TEMPO, BOXPLOT CON SEABORN

```python
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=df['Tempo'], palette='Set1')
ax.grid(False)
ax.set_title('Distribution of Tempo')
plt.show()
```

Esta distribución parece ser aproximadamente normal. La mayoria de las canciones tienen tiempos moderados, pero hay valores atípicos muy grandes o muy chicos
